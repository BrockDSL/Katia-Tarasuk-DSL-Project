We don't want all of these columns in the dataframe, as we are only going to be using specific columns' data, so we're going to pick out the columns that we want to use, and reassign the data that we want to a new variable called df1.

Note: If you would rather erase the the columns than just specify which columns/rows you want to keep, you would use .drop()

Additional Note: if you want to drop a row, you would change the axis=1 to axis=0

```python
df1 = df[['Date/Time','Year','Month','Day','Max Temp (°C)','Min Temp (°C)']]
df1b = df.drop(['Climate ID','Data Quality'],axis=1)
df1.head()
```

Now there is a lot of NaN's (or non-existant data) in the columns that we are going to be using, especially towards the end of the dataframe as this particular set of data was obtained on November 7th, 2019. That being said, we cannot guarantee that there are rows randomly sprinkled throughout the dataframe that also have NaN in place of actual data. So we are going to filter out the rows that have 3 or more NaN's. df.dropna(thresh=n) will give us a dataframe that has at least n definite values, but let's say that we want to filter out rows with at least n NaN's instead of filtering the definite values, so we use df.dropna(thresh=df.shape[1]-n)

In this example we want to filter the rows that have 2 NaN's in the Min and Max Temp Columns, so we can use:

```python
df2 = df1.dropna(subset=['Max Temp (°C)','Min Temp (°C)'], thresh=2)
df2.head()
```

But let's say that we want to filter out everything from November 7th on, and ignore if there are any random other rows with  NaN's. So what we can do is use our and/ors.

Basically using the and/ors, we will filter the data into a new dataframe, df2b, where the number of the month is either less than 11 (November), or the month is 11 and the date is less than the 7th

Note: | means 'or' and & means 'and' in Pandas speech. Regular Python just uses 'and' and 'or.'

```python
df2b = df1[(df.Month < 11)|((df.Month == 11)&(df.Day < 7))]
df2b
```

That being said, if you would rather fill in the data with some number/string, what you can do is use .fillna() where it will input 0's in place of NaN's. If you would rather use another number/string, you would replace the 0 with it.

```python
df2c = df1.fillna(0)
df2c
```

But let's say that there are some outliers in your dataset. Instead of just dropping them, you can replace them with NaN's using the replace() method (if you have a specific number you're looking for). 
Note: np.nan uses the NumPy library, which was imported earlier in the program alongside pandas

```python
df3 = df2.replace([-5.0,-10.0],np.nan)
df3
```

If you just want to replace one specific value at one specific location, you can do:

```python
df3c = df2
df3c.at[4,'Min Temp (°C)'] = np.nan
df3c.iloc[4]
```

Let's assume that we only know that the outlier is at 2019-03-09 and that we want the Max Temp value to change, but we don't know the row number. 

We don't want to go searching through all of the data to pick out that one specific spot. Python can do all of that for us too by using if statements and for-loops.

The piece of code below loops through the dataset via the index, searching for the row where both the Month and Day columns have values equal to what was specified. Then after it changes the value in the column specified to NaN, it leaves the loop using 'break'

```python
df3d = df2
for i in df3d.index: 
    if (df3d.at[i,'Month']==3 and df3d.at[i,'Day']==9):
        df3d.at[i,'Max Temp (°C)'] = np.nan
        break
df3d.iloc[67]
```
